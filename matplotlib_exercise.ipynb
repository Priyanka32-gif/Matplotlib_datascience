```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
ax1.plot(x,y,color='blue')
ax1.plot(x,z,color='green')
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.plot(x,y,color='blue')
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,lw=3,ls='dotted',color='blue')
axes[1].plot(x,z,lw=5,color='red')
fig.tight_layout()
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,lw=4,color='blue')
axes[1].plot(x,z,lw=3,ls='dotted',color='red')
fig.tight_layout()
```
