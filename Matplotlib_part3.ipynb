# Figure Size and DPI (Dot Per Inch)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (3,2), dpi = 100)
#figsize takes tuple of width and height
#dpi is also called pixel per inch
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
# now adding axes on the figure
fig = plt.figure(figsize = (3,2))
ax = fig.add_axes([0,0,1,1])
```

```python
#now changing the figsixe
fig = plt.figure(figsize = (10,2)) #10/2 inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#another way
fig,axes = plt.subplots(figsize=(10,2))
axes.plot(x,y)
```

```python
#adding rows and columns
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# ploting two fig on 1 axes
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label='X cubed')

ax.legend() #used to show the labels name of axes
# we can specify the location of the legend using
# .legend(loc = location code)
```
