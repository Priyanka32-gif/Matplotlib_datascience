# Plot appearance

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue')
```

```python
#in color we can also put rgb hex code
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#000000')
```

```python
#passing line width, 1 is default width
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(y,x,color='purple',linewidth = 5)
```

```python
# to add transprancy in line we use alpha
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='silver',lw=10,alpha=0.7)
```

```python
# we can also add line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7,ls='dotted') # or linestyle='--'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7,linestyle =':')
```

```python
# marker() point x,y point in the line
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7, marker='o') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7, marker='*') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=3,alpha=0.7, marker='+') 
```

```python
# to assign the marker size use markersize
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=1, marker='2',markersize='30') 
```

```python
# marker color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7, marker='o', ms=5, markerfacecolor='blue') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3, marker='o',ms='30',mfc='yellow',
        markeredgecolor='green') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,alpha=0.7, marker='o',ms=30,mfc = 'purple',
       mec = 'green',mew=5) 
```

# plot range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,ls='--') 
ax.set_xlim([0,1]) #x limit, pass lower bound and upper bound
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,ls='--') 
ax.set_ylim([0,5]) #y limit, pass lower bound and upper bound
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,ls='--') 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
