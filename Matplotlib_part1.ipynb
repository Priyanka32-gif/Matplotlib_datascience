It is most popular plotting library for python
It gives you control over every apect of a figure.
It was designed to have a similar feel to MatLab's graphical plotting

```python
import matplotlib.pyplot as plt #importing matplotlib
import numpy as np
import pandas as pd
```

```python
%matplotlib inline 
```

```python
x = np.linspace(0,5,11) # x array
x
```

```python
y = x ** 2 # y array
y
```

```python
# functional method of matplotlib to show plots
plt.plot(x,y)
# in jupyter notebook we don't need to write  plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X Label') #nameing label x
plt.ylabel('Y Label') #nameing label y
plt.title('Title') # title name
```

Multi - plot on a same canvas

```python
plt.subplot(1,2,1) #number of row,column,and plot number
plt.plot(x,y,'r') #ploting color red

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Another method -> Object oriented method
canvas = plt.figure() #this will create blank canvas 
#adding axeses to the canvas
axes = canvas.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
#setting x and y lebel
axes.set_xlabel('x Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
canvas  = plt.figure()
axes1 = canvas.add_axes([0.1,0.1,0.8,0.8])
axes2 = canvas.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('larger plot')
axes2.set_title('Smaller plot')
```
