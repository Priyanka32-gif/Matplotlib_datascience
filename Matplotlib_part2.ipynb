```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
#dividing figure in number of rows and columns
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
# now 3 rows and 3 columns
fig,axes = plt.subplots(nrows = 3, ncols = 3)
```

```python
#for fixing the overlaps use plt.tight_layout()
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
# axes object is just the list of the matlib axes
for current_ax in axes:
    current_ax.plot(x,y)
#using iteration to plot on current axes
```

```python
# we can index in the axes 
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot') #setting title

axes[1].plot(y,x)
axes[1].set_title('Second Plot') #setting title
```
